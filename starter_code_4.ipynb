# Visualize 1.1: using MDS with different distance metrics (side-by-side)

```
from sklearn.manifold import MDS

## use'dissimilarity'='precomputed'; otherwise 'eucledian' distance will be used by default
model = MDS(n_components=2, dissimilarity='precomputed')

## run dimensionality reduction 
reduced_dims1 = model.fit_transform(X1)
reduced_dims2 = model.fit_transform(X2)

## create a subplot with nrow and ncol
nrow = 1
ncol = 2
figs = subplots.make_subplots(rows=nrow, cols=ncol)

## create individual plot first
fig1 = go.Scatter(x = reduced_dims1[:, 0], y = reduced_dims1[:, 1], mode="markers", name="mds_cityblock")
fig2 = go.Scatter(x = reduced_dims2[:, 0], y = reduced_dims2[:, 1], mode="markers", name="mds_euclidean")

## add individual plots to subplot
figs.add_trace(fig1, row=1, col=1) 
figs.add_trace(fig2, row=1, col=2) 

## adjust height and width of subplot
figs.update_layout(
  autosize=False,
  width=600 * ncol,
  height=600 * nrow,
)

# # visualize  subplot
figs.show()
```

# Visualize 1.2: t-SNE with different metrics and perplexity (side-by-side)

```
from sklearn.manifold import TSNE

## use 'metric'='precomputed'; otherwise 'eucledian' distance will be used by default
model1 = TSNE(n_components=2,  metric='precomputed', perplexity = 30)
model2 = TSNE(n_components=2, metric='precomputed', perplexity = 50)

## run dimensionality reduction model
reduced_dims11 = model1.fit_transform(X1)
reduced_dims12 = model1.fit_transform(X2)

reduced_dims21 = model2.fit_transform(X1)
reduced_dims22 = model2.fit_transform(X2)

## create a subplot
nrow = 2
ncol = 2
figs = subplots.make_subplots(rows=nrow, cols=ncol)

## create individual plot
fig11 = go.Scatter(x = reduced_dims11[:, 0], y = reduced_dims11[:, 1], mode="markers", name="tsne_cityblock_prelexity_30")
fig12 = go.Scatter(x = reduced_dims12[:, 0], y = reduced_dims12[:, 1], mode="markers", name="tsne_euclidean_prelexity_30")

fig21 = go.Scatter(x = reduced_dims21[:, 0], y = reduced_dims21[:, 1], mode="markers", name="tsne_cityblock_prelexity_50")
fig22 = go.Scatter(x = reduced_dims22[:, 0], y = reduced_dims22[:, 1], mode="markers", name="tsne_euclidean_prelexity_50")

## add individual plots to subplot
figs.add_trace(fig11, row=1, col=1) 
figs.add_trace(fig12, row=1, col=2) 

figs.add_trace(fig21, row=2, col=1) 
figs.add_trace(fig22, row=2, col=2) 

## adjust height and width of subplot
figs.update_layout(
  autosize=False,
  width=600 * ncol,
  height=600 * nrow,
)

## visualize  subplot
figs.show()
```

## loading + clearning data

```
from pyscagnostics import scagnostics

## make sure all columns are numeric
display(df.head())
```

## Get scores for the data frame

```
# get sgagnostic scores from the library
results = scagnostics(df)
```

## Convert scores to Scagnostic-format

```
# post-process results to make scagnostic

# create 9-attribute list
score_names = ['Outlying', 'Skewed', 'Clumpy', 'Sparse', 'Striated', 'Convex', 'Skinny', 'Stringy', 'Monotonic']

# add a new data source
data = {'score_names':score_names}

scores = None

# iterate over the reasults
for i, j, result in results:
  (measures, bins) = result  
  scores = measures
  
  ## debug
  print(i, j, measures)
  
  col_name = "{}_{}".format(i, j)  
  
  # copy scores
  row = []
  row.append(scores['Outlying'])
  row.append(scores['Skewed'])
  row.append(scores['Clumpy'])
  row.append(scores['Sparse'])
  row.append(scores['Striated'])
  row.append(scores['Convex'])
  row.append(scores['Skinny'])
  row.append(scores['Stringy'])
  row.append(scores['Monotonic'])

  # add current 'col_name' as key and 'row' as value to 'data'
  data[col_name] = row;
# end for
```

## Create a dataframe for scagnostic visualization

```
# convert 'data' to data-frame named 'df_scagnostic'
df_scagnostic = pd.DataFrame.from_dict(data);

# make the first column 'score_names' to index
df_scagnostic.set_index('score_names', inplace=True)

# check out the dataframe
display(df_scagnostic.head(9))
```

## Finally, visualiza scagnostic

```
# now that our dataframe is in perfect condition, visualiza it with plotly
import plotly.express as px

fig = px.imshow(df_scagnostic, text_auto=True, aspect="auto")
fig.update_xaxes(side="top")
fig.show()
```

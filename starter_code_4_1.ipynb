### Preprocessing 1: data cleaning

```
# removing non-numeric, unwanted columns, and rows with missing data
# df = df.drop(['Unnamed: 0', 'city', 'latd', 'longd'], axis = 1)
df = df.drop(['label'], axis = 1)



# removing rows with empty values
df.dropna(inplace=True)
display(df.head())
```

## Create X from the dataframe

```
## making X
X = df.values
```

## Preprocessing 3 : Run different distance metrics
Documentation: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.distance_metrics.html#sklearn.metrics.pairwise.distance_metrics

Try each of the following metric: ['cityblock', 'cosine', 'euclidean', 'haversine']


```
from sklearn.metrics.pairwise import *

# i'm running two metric. but you need to try others to find the best result
X1 = pairwise_distances(X, metric = 'cityblock') # feel free to change
X2 = pairwise_distances(X, metric = 'euclidean') # feel free to change

# X3 = pairwise_distances(df, metric = 'cosine') # feel free to change
# X4 = pairwise_distances(df, metric = 'haversine') # feel free to change

print(X1.shape, X2.shape)
print(X1)
```

# Q.4.1.1: Use MDS with different distance metrics (side-by-side) to find the best visualization that clearly shows 10 separable clusters

```
from sklearn.manifold import MDS

## use'dissimilarity'='precomputed'; otherwise 'eucledian' distance will be used by default
model = MDS(n_components=2, dissimilarity='precomputed')

## run dimensionality reduction 
reduced_dims1 = model.fit_transform(X1)
reduced_dims2 = model.fit_transform(X2)

## create a subplot with nrow and ncol
nrow = 1
ncol = 2
figs = subplots.make_subplots(rows=nrow, cols=ncol)

## create individual plot first
fig1 = go.Scatter(x = reduced_dims1[:, 0], y = reduced_dims1[:, 1], mode="markers", name="mds_cityblock")
fig2 = go.Scatter(x = reduced_dims2[:, 0], y = reduced_dims2[:, 1], mode="markers", name="mds_euclidean")

## add individual plots to subplot
figs.add_trace(fig1, row=1, col=1) 
figs.add_trace(fig2, row=1, col=2) 

## adjust height and width of subplot
figs.update_layout(
  autosize=False,
  width=600 * ncol,
  height=600 * nrow,
)

# # visualize  subplot
figs.show()
```

# Q.4.1.2: Use  t-SNE with different metrics and perplexity (side-by-side) to find the best visualization that clearly shows 10 separable clusters

```
from sklearn.manifold import TSNE

## use 'metric'='precomputed'; otherwise 'eucledian' distance will be used by default
model1 = TSNE(n_components=2,  metric='precomputed', perplexity = 30) # feel free to change
model2 = TSNE(n_components=2, metric='precomputed', perplexity = 50) # feel free to change

## run dimensionality reduction model
reduced_dims11 = model1.fit_transform(X1)
reduced_dims12 = model1.fit_transform(X2)

reduced_dims21 = model2.fit_transform(X1)
reduced_dims22 = model2.fit_transform(X2)

## create a subplot
nrow = 2
ncol = 2
figs = subplots.make_subplots(rows=nrow, cols=ncol)

## create individual plot
fig11 = go.Scatter(x = reduced_dims11[:, 0], y = reduced_dims11[:, 1], mode="markers", name="tsne_cityblock_prelexity_30") 
fig12 = go.Scatter(x = reduced_dims12[:, 0], y = reduced_dims12[:, 1], mode="markers", name="tsne_euclidean_prelexity_30")

fig21 = go.Scatter(x = reduced_dims21[:, 0], y = reduced_dims21[:, 1], mode="markers", name="tsne_cityblock_prelexity_50")
fig22 = go.Scatter(x = reduced_dims22[:, 0], y = reduced_dims22[:, 1], mode="markers", name="tsne_euclidean_prelexity_50")

## add individual plots to subplot
figs.add_trace(fig11, row=1, col=1) 
figs.add_trace(fig12, row=1, col=2) 

figs.add_trace(fig21, row=2, col=1) 
figs.add_trace(fig22, row=2, col=2) 

## adjust height and width of subplot
figs.update_layout(
  autosize=False,
  width=600 * ncol,
  height=600 * nrow,
)

## visualize  subplot
figs.show()
```
